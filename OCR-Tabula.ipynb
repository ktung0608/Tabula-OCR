```python
#import libraries
#%pip install tabula-py in jupyter notebook or install via CLI on local machine
#once completed, restarted the kernel

import tabula
import pandas as pd
import glob

# Define your folder paths

# /User/user/Desktop/
#    |__ file1.pdf
#    |__ file2.pdf
#    |__ template.json

path_to_folder = "path/to/your/folder/"
path_to_template1 = "path/to/template-1.json"
path_to_template2 = "path/to/template-2.json"

# Create output dataframe
output_df = pd.DataFrame()

# for each file in folder ending with .pdf, tabula will read and extract the text into a dataframe
for file in glob.glob(path_to_folder + "*.pdf"):
    
    # You may set how many pages of the PDF you wish to perform the OCR
    for pg in range(1,5):
    

        if pg == 1:
            template = path_to_template1
        else:
            template = path_to_template2
        
        try:
            temp_df = tabula.read_pdf_with_template(file,
                                               template,
                                               pages=str(pg),stream=True)

            # When tabula does not accurately align the credit column, this section will do cleansing to maintain 
            # consistency in the dataframe structure

            if temp_df[0]['Credit'].isnull().all():
                df = temp_df[0].drop('Credit',axis=1)
                df.rename(columns={'Unnamed: 0':'Credit'}, inplace=True)
                df['source'] = str(file)
            else:
                df = temp_df[0]
                df['source'] = str(file)

            # append to the final output table
            output_df = pd.concat([output_df, df], ignore_index=True)
        except:
            pass
        
# export dataframe to destination folder
output_df.to_csv(path_to_folder +'your-filename.csv')
```
